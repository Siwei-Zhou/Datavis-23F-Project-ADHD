```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the image
years = np.array([2005, 2006, 2007, 2008, 2009, 2010])
all_visits = np.array([13379, 16947, 26369, 23735, 26711, 31244])
male_visits = np.array([9059, 10614, 14979, 12287, 14503, 17174])
female_visits = np.array([4315, 6332, 11390, 11448, 12208, 14068])

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(10, 4))

# Plotting the lines
ax.plot(years, all_visits, marker='s', linestyle='-', color='#786028', label='All')
ax.plot(years, male_visits, marker='o', linestyle='-', color='#1984c5', label='Males')
ax.plot(years, female_visits, marker='^', linestyle='-', color='#de6e56', label='Females')

# Adding labels and title
ax.set_xlabel('Year',weight = 'bold')
ax.set_ylabel('Number of ED Visits',weight = 'bold')
ax.set_title('Emergency Department (ED) Visits Related to Attention Deficit/Hyperactivity Disorder (ADHD) Stimulant Medications, by Gender: 2005 to 2010', fontsize=14, weight='bold',pad=30,wrap=True)

# Hide the grid
ax.grid(False)

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Adding legend to the right of the plot
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize = 8)

# Display the plot with a tight layout to ensure the legend is visible
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data from the image
age_groups = ['Aged 5 to 11', 'Aged 12 to 17', 'Aged 18 to 25*', 'Aged 26 to 34*', 'Aged 35 or Older*']
visits_2005 = [3322, 2702, 2131, 1754, 2519]
visits_2010 = [3791, 3461, 8148, 6094, 7957]

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(10, 5))

# Width of a bar
bar_width = 0.35
# Hide the grid
ax.grid(False)

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)




# Position of bars on x-axis
ind = range(len(visits_2005))

# Plotting both years' visits
bar1 = plt.bar(ind, visits_2005, bar_width, label='2005', color='#22a7f0')
bar2 = plt.bar([i + bar_width for i in ind], visits_2010, bar_width, label='2010', color='#e1a692')

# Adding labels and title with specified text sizes
plt.xlabel('Age Group', fontsize=10, weight ='bold')
plt.ylabel('Number of ED Visits', fontsize=10, weight = 'bold')
plt.title('Emergency Department (ED) Visits Related to ADHD Stimulant Medications, by Age Group: 2005 and 2010', fontsize=18, weight='bold',pad=30,wrap=True)

# X-axis Tick labels with increased size
plt.xticks([i + bar_width / 2 for i in ind], age_groups, fontsize=10)

# Adding legend with a specified text size
plt.legend(fontsize=8)

# Display the plot
plt.tight_layout()
plt.show()
```
