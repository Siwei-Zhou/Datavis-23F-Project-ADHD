```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("combined_df_Comorbidity_2016_2022.csv")
df
```

```python
new_column_names = {
    "K2Q30A": "Learning_Disability",
    "K2Q31A": "ADHD",
    "K2Q32A": "Depression",
    "K2Q33A": "Anxiety",
    "K2Q34A": "Behavior_Problems",
    "K2Q35A": "Autism",
    "K2Q36A": "Developmental_Delay",
    "K2Q37A": "Speech_Disorder",
    "K2Q38A": "Tourette_Syndrome",
    "K2Q40A": "Asthma",
    "K2Q42A": "Epilepsy",
    "K2Q43B": "Deafness",
    "K2Q60A": "Intellectual_Disability",
    "K2Q61A": "Cerebral_Palsy"
}

df.rename(columns=new_column_names, inplace=True)
df
```

```python
mapping = {1.0: 'Yes', 2.0: 'No'}

columns_to_map = ["Learning_Disability", "ADHD", "Depression", "Anxiety", "Behavior_Problems", "Autism", "Developmental_Delay", "Speech_Disorder", "Tourette_Syndrome", "Asthma", "Epilepsy", "Deafness", "Intellectual_Disability", "Cerebral_Palsy"]

df[columns_to_map] = df[columns_to_map].replace(mapping)
df
```

```python
nan_counts = df.isna().sum()
print(nan_counts)
```

```python
df['Sum'] = (df == "Yes").sum(axis=1)
df
```

#Prevelance in General Population

```python
disease_columns = df.columns[1:15]
disease_prevalence = {}

for column in disease_columns:
    total_count = len(df)
    yes_count = len(df[df[column] == "Yes"])
    prevalence = (yes_count / total_count) * 100
    disease_prevalence[column] = prevalence

prevalence_df = pd.DataFrame(list(disease_prevalence.items()), columns=["Disease", "Prevalence"])

years = df["Year"].unique()
for year in years:
    year_df = df[df["Year"] == year]
    for column in disease_columns:
        total_count = len(year_df)
        yes_count = len(year_df[year_df[column] == "Yes"])
        prevalence = (yes_count / total_count) * 100 if total_count else 0
        prevalence_df.loc[prevalence_df['Disease'] == column, year] = prevalence

prevalence_df
```

```python
prevalence_df.to_csv('prevalence_df.csv', index=False)
```

# Comorbidity rate of ADHD

```python
adhd_yes_df = df[df['ADHD'] == "Yes"]

disease_columns = adhd_yes_df.columns[1:15]
disease_prevalence = {}

for column in disease_columns:
    total_count = len(adhd_yes_df)
    yes_count = len(adhd_yes_df[adhd_yes_df[column] == "Yes"])
    prevalence = (yes_count / total_count) * 100
    disease_prevalence[column] = prevalence

comorbidity_df = pd.DataFrame(list(disease_prevalence.items()), columns=["Disease", "Prevalence"])

years = adhd_yes_df["Year"].unique()
for year in years:
    year_df = adhd_yes_df[adhd_yes_df["Year"] == year]
    for column in disease_columns:
        total_count = len(year_df)
        yes_count = len(year_df[year_df[column] == "Yes"])
        prevalence = (yes_count / total_count) * 100 if total_count else 0
        comorbidity_df.loc[comorbidity_df['Disease'] == column, year] = prevalence

comorbidity_df
```

```python
comorbidity_df.to_csv('comorbidity_df.csv', index=False)
```

# Rate of Comorbidity Amount

X - Year.   
Y - Comorbidity amount.   
Number - Patient rate.   

```python
years = adhd_yes_df["Year"].unique()
yearly_sum_counts = pd.DataFrame(columns=years)

for year in years:
    year_df = adhd_yes_df[adhd_yes_df["Year"] == year]
    year_sum_counts = year_df['Sum'].value_counts()
    total_counts = year_sum_counts.sum()
    yearly_sum_counts = pd.concat([yearly_sum_counts, year_sum_counts/total_counts], axis=1)

yearly_sum_counts.fillna(0, inplace=True)
yearly_sum_counts.sort_index(inplace=True)
yearly_sum_counts = yearly_sum_counts.iloc[:, -7:]
yearly_sum_counts.columns = [str(year) for year in range(2016, 2023)]
yearly_sum_counts
```

```python
yearly_sum_counts.to_csv('comorbidity_amount.csv', index=False)
```
